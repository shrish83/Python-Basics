```python
train = train.drop(['last_evaluation_bin','average_montly_hours_bin'], axis=1)
```

```python
train.isnull().sum()
```

```python
train1 = train.drop(['satisfaction_level',	'last_evaluation',	'average_montly_hours'],axis=1)
```

```python
for col in train1.columns:  
    train1[col] = train1[col].astype('category') 
    train1[col] = train1[col].cat.codes

train1.head()
```

```python
train2 = train.drop(['number_project','time_spend_company','Work_accident','left','promotion_last_5years','Department','salary'],axis=1)
df = pd.concat([train1,train2],axis=1)
df.head()
```

```python
matrix = df.corr()
f, ax = plt.subplots(figsize = (9,6))
sns.heatmap(matrix, vmax=0.8, square=True, cmap="BuPu")
```

Model Building

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df_train = df.drop(['left'],axis=1)
df_train.head()
```

```python
X_train,X_test,y_train, y_test =  train_test_split(df_train,df.left,train_size=0.75)
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```
