```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets
```

```python
iris = datasets.load_iris()
X = iris.data[:,:2]
y = iris.target
```

```python
#we create an instance of SVM and fit our data. We do not scale our data since we plot the support vectors C=1.0
# SVM regularization parameter
svc = svm.SVC(kernel='linear',C=1,gamma='auto').fit(X,y)
```

```python
#create a mesh to plot in
x_min, x_max = X[:,0].min() - 1, X[:,0].max() + 1
y_min, y_max = X[:,1].min() - 1, X[:,1].max() + 1
h = (x_max/x_min)/100 
xx, yy = np.meshgrid(np.arange(x_min,x_max,h), np.arange(y_min,y_max,h)) 
```

```python
plt.subplot(1, 1, 1)
Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, cmap= plt.cm.Paired, alpha=0.8)
plt.scatter(X[:,0], X[:,1], c=y, cmap=plt.cm.Paired)
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.xlim(xx.min(),xx.max())
plt.title("SVC with linear kernel")
plt.show()
```

SVC with rbf Kernel

```python
svc = svm.SVC(kernel='rbf',C=1,gamma='auto').fit(X,y)

x_min, x_max = X[:,0].min() - 1, X[:,0].max() + 1
y_min, y_max = X[:,1].min() - 1, X[:,1].max() + 1
h = (x_max/x_min)/100 
xx, yy = np.meshgrid(np.arange(x_min,x_max,h), np.arange(y_min,y_max,h)) 

plt.subplot(1, 1, 1)
Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, cmap= plt.cm.Paired, alpha=0.8)
plt.scatter(X[:,0], X[:,1], c=y, cmap=plt.cm.Paired)
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.xlim(xx.min(),xx.max())
plt.title("SVC with rbf kernel")
plt.show()
```

NOTE:::

I would suggest you to go for linear kernel if you have larger number of features(>1000) because it is more likely to be linearly separable in high dimensional space.

Note:::

gamma: Kernel coefficients be 'rbf', 'poly', 'sigmoid'. Higher the value of gamma, will try to exact fit as per training set i.e. generalization error and cause overfitting problem

```python
###Try with different values of gamma

#svc = svm.SVC(kernel='rbf',C=1,gamma=0.10).fit(X,y)
svc = svm.SVC(kernel='rbf',C=1,gamma=100).fit(X,y)


x_min, x_max = X[:,0].min() - 1, X[:,0].max() + 1
y_min, y_max = X[:,1].min() - 1, X[:,1].max() + 1
h = (x_max/x_min)/100 
xx, yy = np.meshgrid(np.arange(x_min,x_max,h), np.arange(y_min,y_max,h)) 

plt.subplot(1, 1, 1)
Z = svc.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, cmap= plt.cm.Paired, alpha=0.8)
plt.scatter(X[:,0], X[:,1], c=y, cmap=plt.cm.Paired)
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.xlim(xx.min(),xx.max())
plt.title("SVC with rbf kernel")
plt.show()
```
