```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
```

```python
sns.set()
sns.set(style='darkgrid')

%matplotlib inline
warnings.filterwarnings("ignore")
```

```python
#Load data
df = pd.read_csv("bigmart_data.csv")
df = df.dropna(how='any')
```

```python
#Line Chart
sns.lineplot(x='Item_Weight',y='Item_MRP',data=df[:50])
```

```python
#barcharts
#sns.barplot

#histogram
#sns.distplot()

#sns.boxplot()

#violin
#sns.violinplot()

#scatter
#sns.relplot(kind='scatter')

#hue plot: var1 vs var2 by coloring with var3
sns.relplot(x='Item_MRP',y='Item_Outlet_Sales',hue='Item_Type' ,data=df[:50])
```

Advance Plots using Seaborn

1. Categorical Scatterplots: One variable is Categorical


a. Strip Plot  : kind =strip

```python
sns.catplot(x='Outlet_Size',y='Item_Outlet_Sales',kind='strip',data=df[:250])
```

b. Swarm Plot: similar to srip but points are adjusted so that they don't overlap

```python
sns.catplot(x='Outlet_Size',y='Item_Outlet_Sales',kind='strip',data=df[:250])
```

Boxplot: categorical

```python
sns.catplot(x='Outlet_Size',y='Item_Outlet_Sales',kind='box',data=df)

##kind= bar for barplot
##kind=violin for violin plot
```

2. Density Plots: uses kernel density estimation 

```python
plt.figure(figsize=(7,7))
sns.kdeplot(df['Item_Visibility'],shade=True)
```

```python
#histogram+density
plt.figure(figsize=(7,7))
sns.distplot(df['Item_Outlet_Sales'])
```

3. Pair Plots: useful to find any correlation in a multidimensional data

```python
data = sns.load_dataset('iris')
data.head()
```

```python
sns.pairplot(data,hue='species', height=2.5)  #height of plot
```
