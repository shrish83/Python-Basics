MatplotLib Tutorial

A plotting library for python and its numerical mathematical extension numPy. It provides an object oriented API for embedding plots into applications using general purpose GUI toolkits like Tkinter, wixPython, Qt or GTK+.

Pros of using this lib:

1. easy to get started
2. Support for custom labels and texts
3. high quality output
4. Customizable
5. Great control of every element in a figure

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#help create different visualization here on notebook it self
#and it removes the need of writing plt.show()
%matplotlib inline
```

```python
height = [111.5,2202.3, 3332,500]
weight = [49,55,83,67]

#line plot
plt.plot(height,weight)
plt.title('height vs weight')
plt.xlabel('height')
plt.ylabel('weight')
```

```python
caloriesBurnt = [210,40,22,145]

plt.plot(caloriesBurnt)
plt.plot(weight)
plt.legend(labels=['Calories Burnt','Weight'],loc='lower right')
```

```python
plt.plot(caloriesBurnt)
plt.plot(weight)
plt.legend(labels=['Calories Burnt','Weight'],loc='lower right')

plt.xticks(ticks=[0,1,2,3], labels=['p1','p2','p3','p4'])
```

```python
#size
plt.figure(figsize=[15,5])

#color
plt.plot(weight,'y--') #y->yellow, -- -> dashed line
plt.plot(caloriesBurnt,'go') #g -> green, o -> as a circle
```

```python
df = pd.read_csv("bigmart_data.csv")
df.head()
```

LINE CHARTS  - default plot type

```python
pricebyitem = df.groupby('Item_Type').Item_MRP.mean()
pricebyitem
```

```python
x = pricebyitem.index.tolist()
y = pricebyitem.values.tolist()

plt.figure(figsize=(14,8))
plt.plot(x,y)
```

BAR CHART

```python
pricebyoutlet = df.groupby('Outlet_Size').Item_Outlet_Sales.mean()
pricebyoutlet
```

```python
x = pricebyoutlet.index.tolist()
y = pricebyoutlet.values.tolist()

plt.bar(x,y, color=['Red','Blue','Green'])
```

HISTOGRAMS

```python
plt.hist(df['Item_MRP'], bins=20, color='Pink')
```

BOX PLOTS

```python
data = df[['Item_Outlet_Sales']]
plt.boxplot(data.values, labels=['Item_Sales'])
```
