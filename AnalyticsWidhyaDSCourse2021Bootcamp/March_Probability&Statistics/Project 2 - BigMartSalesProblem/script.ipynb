```python
bins = [5,10,15,20]
groups = ['Low','Average','High']
train['ItemWeight_bin'] = pd.cut(train['Item_Weight'],bins,labels=groups)

train.groupby('ItemWeight_bin')['Item_Outlet_Sales'].mean().plot.bar()
```

```python
bins = [0.00,0.06,0.12,0.18,0.24,0.30]
groups = ['Very Low','Low','Average','High','Very High']
train['Item_Visibility_bin'] = pd.cut(train['Item_Visibility'],bins,labels=groups)

train.groupby('Item_Visibility_bin')['Item_Outlet_Sales'].mean().plot.bar()
```

```python
bins = [25,70,135,200,260]
groups = ['Low','Average','High','Very High']
train['Item_MRP_bin'] = pd.cut(train['Item_MRP'],bins,labels=groups)

train.groupby('Item_MRP_bin')['Item_Outlet_Sales'].mean().plot.bar()
```

```python
bins = [1980,1990,2000,2010]
groups = ["early80","early90","early20"]
train['Outlet_Establishment_Year_bin'] = pd.cut(train['Outlet_Establishment_Year'],bins,labels=groups)

train.groupby('Outlet_Establishment_Year_bin')['Item_Outlet_Sales'].mean().plot.bar()
```
