MODEL BUILDING

```python
from pandas import Series
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

lreg = LinearRegression()
```

```python
train.head(20)
```

```python
#Converting visibity=0 with mean
train['Item_Visibility'] = train['Item_Visibility'].replace(0,np.mean(train['Item_Visibility']))

#Converting Outlet_Establishment_Year
train['Outlet_Establishment_Year'] = 2013 - train['Outlet_Establishment_Year']

train['Outlet_Size'].fillna('Small', inplace=True)
```

```python
train1 = train.drop(['Item_Weight','Item_Visibility',	'Item_MRP'	,'Outlet_Establishment_Year',	'Item_Outlet_Sales','Item_Visibility_log',	'Item_Outlet_Sales_log'],axis=1)
```

```python
train1= train1.drop(['Item_Identifier','Outlet_Identifier'],axis=1)
train1.head()
```

```python
dummies = pd.get_dummies(train1)
dummies.head()
```

```python
train2 = train.drop(['Item_Fat_Content','Item_Type','Outlet_Size','Outlet_Location_Type','Outlet_Type','Item_Identifier','Outlet_Identifier'],axis=1)
X = pd.concat([train2,dummies], axis=1)
```

```python
X = X.drop(['Item_Outlet_Sales','Item_Outlet_Sales_log'],1)
```

```python
x_train, x_cv, y_train, y_cv = train_test_split(X,train.Item_Outlet_Sales, test_size=0.3)
```

```python
lreg.fit(x_train,y_train)
```

```python
#predict on cv
pred_cv = lreg.predict(x_cv)
```

```python
#mse
mse = np.mean((pred_cv - y_cv) ** 2)
mse
```

```python
#r square
lreg.score(x_cv,y_cv)
```

Residual Vs fitted Values Plot

```python
x_plot = plt.scatter(pred_cv, (pred_cv- y_cv), c='b')
plt.hlines(y=0, xmin=-1000, xmax=5000)
plt.title("Residual Plot")
```

We see a funnel shape in the plot. This shape is called "Heteroskedasticity". It is caused by the presence of non linearity in data. We can clearly see the variance of error terms(residuals) is not constant. This variance is because of outliers or extrememly leverage values. These values get too much weight thereby disproportionaltely affecting the model's performance. Hence, we look out for the solution to "POLYNOMIAL REGRESSION" .
